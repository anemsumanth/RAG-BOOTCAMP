```python
from langchain_community.document_loaders import UnstructuredPDFLoader
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

class SmartPDFProcessor:
    """advanced PDF processing class using UnstructuredPDFLoader and RecursiveCharacterTextSplitter."""
    def __init__(self, file_path):
        self.file_path = file_path
        self.loader = UnstructuredPDFLoader(file_path)
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200,
            separators=[""]
        )

    def load_and_split(self):
        try:
            documents = self.loader.load()
            print(f"Loaded {len(documents)} documents from {self.file_path}")
            split_docs = self.text_splitter.split_documents(documents)
            print(f"Split into {len(split_docs)} chunks")
            return split_docs
        except Exception as e:
            print(f"Error processing PDF: {e}")
            return []
# Example usage
processor = SmartPDFProcessor("data/pdf/SearchOptimization.pdf")
```

```python
from langchain_core.documents import Document
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

print("Loading PDF using PDFLoader...")

class PDFLoader:
    def __init__(self, file_path):
        self.file_path = file_path
        self.loader = PyPDFLoader(file_path)
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200,
            length_function=len
        )

    def load_and_split(self):
        try:
            documents = self.loader.load()
            
            print(f"Loaded {len(documents)} documents from {self.file_path}")
            for doc_num, doc  in enumerate(documents):
               
               print(f"Split into {len(documents)} chunks")
               print(documents[0].page_content[:500])  # Print first 500 characters of the first document
               if len(documents) > 1:
                   print(f"Document {doc_num + 1} has {len(documents[doc_num].page_content)} characters.")
            chuncks    = self.text_splitter.create_documents(doc.page_content)
            return chuncks
        except Exception as e:
            print(f"Error processing PDF: {e}")
            return []

# Example usage
pdf_loader = PDFLoader("data/pdf/SearchOptimization.pdf")
print(pdf_loader.load_and_split())
```
